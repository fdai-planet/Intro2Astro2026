## Exercise 1 — Meteor activity through one night

Plot the hourly meteor counts as a dark-orange line with circle markers. Add axis labels, the title **Meteor activity during one night**, a grid, and an x-range covering the observations.

```python
# Provided data — run this cell before working on the exercise.
hours = np.arange(-4, 8)
counts = np.array([12, 18, 35, 62, 95, 121, 88, 54, 31, 20, 14, 9])
```

### Answer

```python
fig, ax = plt.subplots(figsize=(7, 4))  # Create the canvas and one plotting area.
ax.plot(hours, counts, color="darkorange", marker="o", lw=2)
ax.set(
    xlabel="Hours from midnight",
    ylabel="Meteors per hour",
    title="Meteor activity during one night",
    xlim=(hours.min(), hours.max()),
)
ax.grid(alpha=0.3)
fig.tight_layout()  # Prevent labels from being clipped.
plt.show()
```

## Exercise 2 — Compare two stellar spectra

Plot the quiet star as a navy solid line and the active star as a crimson dashed line with width 2. Add a frameless legend, minor ticks, inward-facing ticks, a light grid, and suitable labels.

```python
# Provided data — the formulas only create two example spectra.
wavelength = np.linspace(654, 658, 300)
quiet = 1 - 0.38 * np.exp(-0.5 * ((wavelength - 656.28) / 0.18) ** 2)
active = 1 - 0.22 * np.exp(-0.5 * ((wavelength - 656.28) / 0.32) ** 2)
```

### Answer

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(wavelength, quiet, color="navy", label="Quiet star")
ax.plot(wavelength, active, color="crimson", ls="--", lw=2, label="Active star")
ax.set(xlabel="Wavelength (nm)", ylabel="Normalized flux", ylim=(0.55, 1.04))
ax.minorticks_on()
ax.tick_params(which="both", direction="in")
ax.grid(alpha=0.2)
ax.legend(frameon=False)
fig.tight_layout()
plt.show()
```

## Exercise 3 — Annotate an atmospheric profile

Plot atmospheric pressure against altitude using a logarithmic pressure axis. Annotate the measurement at 10 km with an arrow labelled **about one quarter of sea-level pressure**, and place the text **pressure falls rapidly with altitude** near the top of the plot.

```python
# Provided data — pressure measured at seven altitudes.
altitude = np.arange(0, 31, 5)
pressure = np.array([1013, 540, 265, 120, 55, 25, 12])
```

### Answer

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(pressure, altitude, "o-", color="tab:purple")
ax.set(xscale="log", xlabel="Pressure (hPa)", ylabel="Altitude (km)")
ax.annotate(
    "about one quarter of sea-level pressure",
    xy=(265, 10),  # The data position identified by the arrow.
    xytext=(45, 16),  # The position of the annotation text.
    arrowprops={"arrowstyle": "->"},
)
ax.text(18, 27, "pressure falls rapidly with altitude", color="0.35")
ax.grid(alpha=0.3, which="both")
fig.tight_layout()
plt.show()
```

## Exercise 4 — Find the best telescope focus

Plot the focus-scan samples with a purple dash-dot line and diamond markers. Highlight the minimum with a larger gold star and a legend. Include commands that would save the result as a 300 dpi PNG and as a vector PDF.

```python
# Provided data — smaller FWHM means a sharper stellar image.
focus = np.linspace(-2, 2, 17)
fwhm = 1.3 + 0.45 * (focus - 0.3) ** 2
best = np.argmin(fwhm)
```

### Answer

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(focus, fwhm, color="tab:purple", ls="-.", marker="D", ms=5, label="Focus scan")
ax.scatter(focus[best], fwhm[best], marker="*", s=180, color="gold", edgecolor="black", label="Best focus", zorder=3)
ax.set(xlabel="Focus position (mm)", ylabel="Stellar FWHM (pixels)")
ax.legend(frameon=False)
fig.tight_layout()
# Save before plt.show(); uncomment these lines when you want files.
# fig.savefig("focus-scan.png", dpi=300)
# fig.savefig("focus-scan.pdf")
plt.show()
```
