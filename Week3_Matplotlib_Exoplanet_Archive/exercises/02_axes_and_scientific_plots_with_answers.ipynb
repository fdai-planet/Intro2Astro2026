```python
import numpy as np
import matplotlib.pyplot as plt
# if you want to drag the figure, install ipympl and use it by uncommenting the next line
# %matplotlib ipympl

rng = np.random.default_rng(42)
```

## Exercise 1 — Transit data and residuals

Create two vertically stacked Axes sharing the x-axis. In the upper panel, show the noisy transit observations as small black points and the model as an orange line. In the lower panel, plot the residuals, add a horizontal zero line, and label both panels.

```python
# Provided data — run this cell before working on the plot.
time = np.linspace(-0.12, 0.12, 121)
model = 1 - 0.018 * np.exp(-(time / 0.045) ** 8)
flux = model + rng.normal(0, 0.0015, time.size)
residual = flux - model
```

### Answer

```python
fig, (ax_flux, ax_residual) = plt.subplots(
    2, 1, figsize=(7, 6), sharex=True, gridspec_kw={"height_ratios": [3, 1]}
)
ax_flux.scatter(time, flux, s=12, color="black", alpha=0.65, label="Observation")
ax_flux.plot(time, model, color="tab:orange", lw=2, label="Transit model")
ax_flux.set(ylabel="Relative flux", title="Synthetic transit observation")
ax_flux.legend(frameon=False)

ax_residual.scatter(time, residual, s=10, color="tab:blue")
ax_residual.axhline(0, color="0.4", ls="--", lw=1)  # Reference for a perfect fit.
ax_residual.set(xlabel="Time from mid-transit (days)", ylabel="Residual")

fig.tight_layout()
plt.show()
```

## Exercise 2 — A radio spectrum with asymmetric errors

Plot the radio measurements with asymmetric vertical error bars, circle markers, and caps. Use logarithmic scales on both axes and add scientific labels. The lower and upper uncertainties are deliberately different.

```python
# Provided data — the two rows contain lower and upper uncertainties.
frequency = np.array([1, 2, 5, 10, 20, 40])
flux_density = np.array([120, 78, 43, 26, 15, 8])
lower_error = np.array([8, 6, 4, 3, 2, 1])
upper_error = np.array([10, 7, 5, 3.5, 2.5, 1.5])
asymmetric_error = np.vstack([lower_error, upper_error])
```

### Answer

```python
fig, ax = plt.subplots(figsize=(6, 4))
ax.errorbar(
    frequency,
    flux_density,
    yerr=asymmetric_error,  # Row 1 is lower error; row 2 is upper error.
    fmt="o",
    capsize=4,
    color="tab:red",
)
ax.set(
    xscale="log",
    yscale="log",
    xlabel="Frequency (GHz)",
    ylabel="Flux density (mJy)",
    title="Radio continuum spectrum",
)
ax.grid(alpha=0.25, which="both")
fig.tight_layout()
plt.show()
```

## Exercise 3 — Telescope throughput and calibration uncertainty

Plot the non-linear throughput curve and shade its calibration interval from `throughput - uncertainty` to `throughput + uncertainty`. Label both elements in a frameless legend and keep the y-range between 0 and 1.

```python
# Provided data — wavelength, throughput, and calibration uncertainty.
wavelength = np.linspace(300, 1000, 300)
throughput = 0.86 * np.exp(-((wavelength - 650) / 250) ** 4)
uncertainty = 0.025 + 0.025 * np.abs(wavelength - 650) / 350
```

### Answer

```python
fig, ax = plt.subplots(figsize=(7, 4))
ax.plot(wavelength, throughput, color="tab:green", lw=2, label="Measured throughput")
ax.fill_between(
    wavelength,
    throughput - uncertainty,
    throughput + uncertainty,
    color="tab:green",
    alpha=0.25,  # Transparency keeps the central curve visible.
    label="Calibration interval",
)
ax.set(xlabel="Wavelength (nm)", ylabel="Throughput", ylim=(0, 1))
ax.legend(frameon=False)
fig.tight_layout()
plt.show()
```

## Exercise 4 — Photon energies and observing conditions

Create a 1 × 2 figure.

1. Left: draw a histogram of the simulated photon energies. Use 30 bins and a step outline.
2. Right: scatter signal-to-noise ratio against exposure time. Color the points by sky brightness and add a colorbar.

```python
# Provided data — no knowledge of these simulation formulas is required.
photon_energy = rng.gamma(shape=2.5, scale=0.8, size=2000)
exposure = rng.uniform(10, 600, 300)
sky_brightness = rng.uniform(0.5, 3.0, 300)
snr = 2.5 * np.sqrt(exposure / sky_brightness) + rng.normal(0, 1.5, 300)
```

### Answer

```python
fig, (ax_hist, ax_scatter) = plt.subplots(1, 2, figsize=(11, 4.5))

ax_hist.hist(photon_energy, bins=30, histtype="step", lw=2, color="tab:purple")
ax_hist.set(xlabel="Photon energy (keV)", ylabel="Count", title="Detected photon energies")

points = ax_scatter.scatter(
    exposure, snr, c=sky_brightness, s=24, cmap="cividis", alpha=0.8
)
ax_scatter.set(xlabel="Exposure time (s)", ylabel="Signal-to-noise ratio", title="Observing conditions")
fig.colorbar(points, ax=ax_scatter, label="Sky brightness")

fig.tight_layout()
plt.show()
```
